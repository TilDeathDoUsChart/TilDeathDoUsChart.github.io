```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lci_data = pd.read_csv("LCI_Final.csv")
```

```python
lci_data.head()
```

```python
ml_results = pd.DataFrame()
ml_results["Count"] = 0
ml_results["Variable"] = 0
ml_results["r2"] = 0
ml_results["mse"] = 0



ml_results["Count"] = ml_results["Count"].astype(object)
ml_results["Variable"] = ml_results["Variable"].astype(object)
ml_results["r2"] = ml_results["r2"].astype(object)
ml_results["mse"] = ml_results["mse"].astype(object)
```

```python
ml_results
```

```python
count = 0
for filename in os.listdir("csv_folder/"):
    #pull in specific variables for the DV
    dv = pd.read_csv("csv_folder/" + filename)
    dv = dv.sort_values(by = ["R0000100"])
    for column in dv:
        try:
            dv_temp = pd.DataFrame()
            dv_temp["PUBID"] = lci_data["PUBID"]
            dv_temp["LCI"] = lci_data["LCI"]
            dv_temp[column] = dv[column]
            dv_temp = dv_temp[dv_temp[column] >=0]
            #build the ML scheme
            X = dv_temp["LCI"].values.reshape(-1,1)
            y = dv_temp[column].values.reshape(-1,1)
            #make training data
            X_train, X_test, y_train, y_test = train_test_split(X, y, random_state =42)
            #model
            model = LinearRegression()
            model.fit(X_train, y_train)
            predicted = model.predict(X_test)
            mse = mean_squared_error(y_test, predicted)
            r2 = r2_score(y_test, predicted)
            #add to table
            ml_results.at[count, "Count"] = count
            ml_results.at[count, "Variable"] = column
            ml_results.at[count, "r2"] = r2
            ml_results.at[count, "mse"] = mse
            ml_results.at[count, "observations"] = len(dv_temp)
            count = count+1
        except(ValueError):
            print(column + " has no data")
```

```python
ml_results.to_csv("ml_results.csv")
```
